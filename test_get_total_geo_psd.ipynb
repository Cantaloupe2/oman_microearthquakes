# Test computing the total PSD of a geophone station

```python
# Imports
from os.path import join
from pandas import Timestamp

from utils_basic import SPECTROGRAM_DIR as indir
from utils_spec import assemble_spec_filename, read_geo_spectrograms
from utils_plot import plot_long_term_geo_stft_spectrograms, save_figure
```

```python
# Inputs
range_type = "daily"
station = "A01"
day = "20200113"
window_length = 60.0
overlap = 0.0
downsample = True
downsample_factor = 60

dbmin = -30.0
dbmax = 0.0

marker = True

starttime_marker = Timestamp("2020-01-13T20:00:00")
endtime_marker = Timestamp("2020-01-13T21:00:00")
```

```python
# Read the spectrograms

# Downsampled
filename = assemble_spec_filename(range_type, "geo", day, station, window_length, overlap, downsample, downsample_factor = downsample_factor)
inpath = join(indir, range_type, filename)
stream_spec_ds = read_geo_spectrograms(inpath)
```

```python
# Compute the total PSD
trace_spec_total_ds = stream_spec_ds.get_total_power()
```

```python
# Plotting
fig, axes, cbar = plot_long_term_geo_stft_spectrograms(stream_spec_ds,
                                                        dbmin = dbmin, dbmax = dbmax, 
                                                        date_format = "%Y-%m-%dT%H:%M:%S", major_time_spacing = 6, minor_time_spacing = 1,
                                                        time_tick_rotation = 5,
                                                        plot_total_psd = True, total_psd_trace = trace_spec_total_ds)
                        
                                    
                                
```
