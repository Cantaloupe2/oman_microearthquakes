# Correct the channel names of the geophone data
## The geophone MSEED data I got from Rob somehow have wrong channel names, which affect the sensitivity removal.

```python
# Imports
from os import walk
from os.path import join
from obspy import read

from utils_basic import ROOTDIR_GEO
```

```python
# Inputs
rootdir = ROOTDIR_GEO
```

```python
# Change the channel names
for rootdirs, dirnames, filenames in walk(rootdir):
    for filename in filenames:
        if filename.endswith('.mseed'):
            inpath = join(rootdirs, filename)
            print(f"Working on {inpath}...")

            stream = read(inpath)
            for trace in stream:
                channel_old = trace.stats.channel
                if channel_old.startswith('B'):
                    channel_new = channel_old.replace('B', 'G')
                    trace.stats.channel = channel_new

                    print(f"Changing channel name from {channel_old} to {channel_new}...")
            
            stream.write(inpath, format='MSEED')
            
            
```
