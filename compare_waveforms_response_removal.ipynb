# Compare the geophone waveforms with and without instrument response removal
## I conduct this comparison because the old instrument-response removal procedure could have introduced noise on the two ends of the spectrum

```python
# Imports
from obspy import read, UTCDateTime
from matplotlib.pyplot import subplots
```

```python
# Inputs
inpath_w = "/Volumes/OmanData/data/geophones_old/2020-01-16-22-00-00/7F.A04.2020-01-16-22-00-00..GHZ.SAC"
inpath_wo = "/Volumes/OmanData/data/geophones/2020-01-16/7F.A04..GHZ.mseed"

starttime = UTCDateTime("2020-01-16T22:00:00")
endtime = UTCDateTime("2020-01-16T22:02:00")
```

```python
# Read
stream_w = read(inpath_w, starttime=starttime, endtime=endtime)
stream_wo = read(inpath_wo, starttime=starttime, endtime=endtime)
```

```python
# Plot
fig, ax = subplots(2, 1, sharex=True, figsize=(10, 5))

trace_w = stream_w[0]
trace_wo = stream_wo[0]

timeax_w = trace_w.times()
timeax_wo = trace_wo.times()

ax[0].plot(timeax_w, trace_w.data, color="black")
ax[1].plot(timeax_wo, trace_wo.data, color="black")
```
