```python
# Make plot of one hour of waveform

from obspy import read
```

```python
inpath = "/Volumes/OmanData/geophones_no_prefilt/data/2020-01-25-12-00-00/7F.A03.2020-01-25-12-00-00..GH*"
stream = read(inpath)
```

```python
trace_z = stream.select(channel="*Z")[0]
trace_1 = stream.select(channel="*1")[0]
trace_2 = stream.select(channel="*2")[0]
```

```python
trace_z.plot()
trace_1.plot()
trace_2.plot()
```

```python
print(trace_1.stats.sac.cmpaz)
print(trace_2.stats.sac.cmpaz)
```
