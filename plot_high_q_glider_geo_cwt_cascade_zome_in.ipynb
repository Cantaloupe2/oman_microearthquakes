# Plot the waveforms and spectrograms computed using CWT in successively zomed-in time-frequency windows

```python
# Imports
from pandas import Timestamp, Timedelta
from utils_basic import ROOTDIR_GEO as rootdir
from utils_basic import time2filename
from utils_wavelet import get_scales_and_freqs, get_stream_cwt
from utils_preproc import read_and_process_windowed_geo_waveforms
from utils_plot import plot_cascade_zoom_in_geo_waveforms_and_cwt_powers, save_figure
```

```python
# Inputs
## Data
starttimes = ["2020-01-13T20:00:00Z", "2020-01-13T20:01:01.5Z"]
durs = [120.0, 10.0]
stations_to_plot = ["A05", "A15", "A19", "B04", "B07", "B20"]
component_to_plot = "Z"

## Preprocessing
min_freq_filt = 40.0
max_freq_filt = 70.0
deci_factor = 2

## Wavelet transform
wavelet = "cmor"
center_freq = 2.0
bandwidth = 20.0
min_scales = 5
max_sacale = 50
num_scales = 128

## Plotting
xdim = 2.0
ydim = 1.5

ylim_wf = (-50.0, 50.0)

min_freqs_plot = [30.0, 30.0]
max_freqs_plot = [200.0, 100.0]

major_time_spacings = [60.0, 5.0]
minor_time_spacings = [15.0, 1.0]

major_freq_spacings = [25.0, 10.0]
minor_freq_spacings = [5.0, 2.0]

dbmin = 0.0
dbmax = 20.0

linewidth_wf = 0.1
```

```python
# Read and preprocess waveforms
## Unfiltered
starttime = starttimes[0]
dur = durs[0]
stream_no_filt = read_and_process_windowed_geo_waveforms(starttime, dur, stations=stations_to_plot, components=component_to_plot, decimate=True, decimate_factor=deci_factor)

## Filtered
stream_filt = read_and_process_windowed_geo_waveforms(starttime, dur, stations=stations_to_plot, components=component_to_plot, freqmin=min_freq_filt, freqmax=max_freq_filt, decimate=True, decimate_factor=deci_factor)
```

```python
# Compute the CWT
scales, freqs = get_scales_and_freqs(wavelet, center_freq, bandwidth, min_scales, max_sacale, num_scales)
specs = get_stream_cwt(stream_no_filt, wavelet=wavelet, scales=scales, bandwidth=bandwidth, center_freq=center_freq)
```

```python
# Plot the waveforms and CWT powers
window_dict = {"starttimes": starttimes, "durations": durs, "min_freqs": min_freqs_plot, "max_freqs": max_freqs_plot, "major_time_spacings": major_time_spacings, "minor_time_spacings": minor_time_spacings, "major_freq_spacings": major_freq_spacings, "minor_freq_spacings": minor_freq_spacings}

fig, axes, cbar = plot_cascade_zoom_in_geo_waveforms_and_cwt_powers(stream_filt, specs, window_dict, min_freq_filt, max_freq_filt, dbmin=dbmin, dbmax=dbmax)
```

```python
# Save the figure
starttime = starttimes[0]
timestr = time2filename(starttime)
figname = f"high_q_glider_cascade_geo_waveforms_and_cwt_powers_{timestr}_{component_to_plot.lower()}.png"
save_figure(fig, figname, outdir=rootdir)
```
