```python
# Downsample the 1D Vs profile
```

```python
## Import modules
import numpy as np
import matplotlib.pyplot as plt
from model_utils import adaptive_downsample
```

```python
## Inputs
ddep_min = 1.0 # minimum depth spacing
ddep_max = 5.0 # maximum depth spacing
thresh = 10 # threshold for downsample

inpath = "/Volumes/OmanData/geophones_no_prefilt/data/shallow_vs_profile.dat"
```

```python
## Read in the 1D Vs profile
model_in = np.loadtxt(inpath)

dep_in = model_in[:, 0]
vel_in = model_in[:, 1]
```

```python
## Perform adaptive downsampling
dep_out, vel_out = adaptive_downsample(dep_in, vel_in, thresh, ddep_min, ddep_max)
```

```python
## Plot the results
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.plot(vel_in, dep_in, 'k-', label='Original')
ax.scatter(vel_out, dep_out, c='r', label='Downsampled')
ax.invert_yaxis()
```

```python
## Save the results
outpath = inpath.replace('.dat', '_downsampled.dat')
np.savetxt(outpath, np.c_[dep_out, vel_out], fmt='%1.3f')
```
