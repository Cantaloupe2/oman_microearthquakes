# Plot the hourly spectrograms of all locations of a hydrophone station

```python
from os.path import join

from utils_basic import SPECTROGRAM_DIR as indir
from utils_spec import read_hydro_spectrograms
from utils_plot import plot_long_term_hydro_stft_spectrograms, save_figure
```

```python
station = "B00"
hour = "20200113200000"
window_length = 1.0
overlap = 0.0

dbmin = -90.0
dbmax = -50.0

indir = join(indir, "hourly")
```

```python
filename = f"hourly_hydro_spectrograms_{hour}_{station}_window{window_length:.0f}s_overlap{overlap:.1f}.h5"
inpath = join(indir, filename)
stream_spec = read_hydro_spectrograms(inpath)
```

```python
fig, axes, cbar = plot_long_term_hydro_stft_spectrograms(stream_spec, 
                                 dbmin = dbmin, dbmax = dbmax, 
                                 date_format = "%Y-%m-%dT%H:%M:%S", major_time_spacing = 0.25, minor_time_spacing = 0.05,
                                 time_tick_rotation = 5)
```

```python
figname = f"hourly_hydro_spectrograms_{hour}_{station}_window{window_length:.0f}s_overlap{overlap:.1f}.png"
save_figure(fig, figname)
```
