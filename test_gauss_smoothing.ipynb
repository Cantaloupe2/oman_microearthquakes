# Test smoothing a matrix using a combination of 1D and 2D convolutions

```python
# Import
from numpy import cos, geomspace, linspace, meshgrid, pi
from numpy.random import randn
from scipy.signal import convolve2d
from matplotlib.pyplot import subplots, colorbar
from utils_wavelet import smooth_along_time, smooth_along_freq, get_gauss_kernel
```

```python
boxcar_window = 3
scales = linspace(1, 100, 100)
times = linspace(0, 1, 1001)

time_period = 0.1
scale_period = 10
```

```python
time_grid, scale_grid = meshgrid(times, scales)
data = cos(time_grid / time_period * 2 * pi) * cos(scale_grid / scale_period * 2 * pi)
data_time_smoothed = smooth_along_time(data, scales)
data_freq_smoothed = smooth_along_freq(data, times, 11)
```

```python
fig, axes = subplots(4, 1, figsize=(10, 10))
axes[0].imshow(data, aspect='auto', vmin=-1, vmax=1)
axes[1].imshow(data_time_smoothed, aspect='auto', vmin=-1, vmax=1)
axes[2].imshow(data_freq_smoothed, aspect='auto', vmin=-1, vmax=1)
```
