```python
from obspy import read
import matplotlib.pyplot as plt
import numpy as np
from os.path import join
```

```python
timewin = "2020-01-14-07-00-00"
sacname = "7F.A01.2020-01-14-07-00-00..GHZ.SAC"

root1 = "/Volumes/OmanData/geophones_no_prefilt/data"
root2 = "/Volumes/OmanData/geophones/data"

begin = 46*60
end = 47*60
```

```python
inpath1 = join(root1, timewin, sacname)
inpath2 = join(root2, timewin, sacname)

stream1 = read(inpath1)
stream2 = read(inpath2)

trace1 = stream1[0]
trace2 = stream2[0]

# starttime = trace1.stats.starttime

# trace1.trim(starttime+begin, starttime+end)
# trace2.trim(starttime+begin, starttime+end)

# trace1.spectrogram(log=True)
# trace2.spectrogram(log=True)

```

```python
data1 = trace1.data
data2 = trace2.data
timeax = trace1.times()

data1 = data1/np.max(data1)
data2 = data2/np.max(data2)

plt.figure(figsize=(10, 5)) 
plt.plot(timeax, data1, label="no_prefilt", color="blue", linewidth=0.2)
plt.plot(timeax, data2, label="prefilt", color="black", linewidth=0.2)
plt.xlim(begin, end)
plt.ylim(-1, 1)
plt.legend()
```
