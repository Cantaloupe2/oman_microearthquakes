# Compute the whole deployment spectrograms for all stations

```python
# Imports
from os.path import join

from utils_basic import GEO_STATIONS as stations, ROOTDIR_GEO as rootdir
from utils_basic import get_geophone_days
from utils_spec import get_whole_deployment_spectrogram, save_geo_spectrogram
```

```python
# Inputs
window_length = 15.0 # IN MINUTES
overlap = 0.5
downsample_factor = 1000 # Downsample factor for the frequency axis
```

```python
# Get days of the geophone deployment
days = get_geophone_days()
```

```python
# Compute the spectrogram for each station and save it
for station in stations:
    # Compute the spectrogram
    specs = get_whole_deployment_spectrogram(station, days, window_length = window_length, overlap = overlap, downsample_factor = downsample_factor)

    # Save the spectrogram
    filename = f"whole_deployment_spectrogram_{station}_{window_length:.0f}mins.npz"
    outpath = join(rootdir, "spectrograms", filename)

    save_geo_spectrogram(specs, outpath)
```
