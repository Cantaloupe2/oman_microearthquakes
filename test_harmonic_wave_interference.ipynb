# Generate examples of two 1D harmonic waves interfering with each other 

```python
# Imports
from numpy import cos, linspace, meshgrid, pi
from matplotlib.pyplot import subplots
```

```python
# Inputs
omega = 2 * pi * 100 # rad/s
c = 1000 # m/s

a1 = 1
a2 = 2

phi = 0

xmin = -100
xmax = 100

tmin = 0
tmax = 0.1

numpts_time = 501
numpts_space = 501
```

```python
# Generate the signals
time_vect = linspace(tmin, tmax, numpts_time)
space_vect = linspace(xmin, xmax, numpts_space)

time_grid, space_grid = meshgrid(time_vect, space_vect)
wave1 = a1 * cos(omega * (time_grid +  space_grid / c) + phi)
wave2 = a2 * cos(omega * (time_grid - space_grid / c))

wave = wave1 + wave2
```

```python
# Plot the signals
fig, axes = subplots(1, 3, figsize=(10, 10), sharex=True, sharey=True)

ax = axes[0]
ax.pcolor(time_grid, space_grid, wave1, cmap="RdBu", shading="auto")

ax = axes[1]
ax.pcolor(time_grid, space_grid, wave2, cmap="RdBu", shading="auto")

ax = axes[2]
ax.pcolor(time_grid, space_grid, wave, cmap="RdBu", shading="auto")
```
